```python
# checking how many of types of flower data

folders = os.listdir(r"H:\innomatics\ML\ML Class\flowers_image_data")  # Whenever we see names it is a folder and when we get extension it is a file
folders
```

```python
# Looping through each image folder and images to convert the images into array
# after conversion of array we resize the array and then performing flatten

features = []
class_labels = []
for folder in folders:
    for images in os.listdir(r"H:\innomatics\ML\ML Class\flowers_image_data\{}".format(folder)):
        img = cv2.imread(r"H:\innomatics\ML\ML Class\flowers_image_data\{}\{}".format(folder,images),0)
        img = cv2.resize(img, (50,50))    # Resizing the Image
        img = img.flatten()    # Flattening tha Image
        features.append(img)
        class_labels.append(folder)
```

```python
# check whether teh length of features and class_labels are same or not

print(len(features))
print(len(class_labels))
```

```python
# using features and class labels creating a final data frame 
# As features and class_labels both are in array we can give it to dataframe

final_data = pd.DataFrame(features)
```

```python
final_data["class_labels"] = class_labels
```
